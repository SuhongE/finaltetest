```python
import matplotlib.pyplot as plt # matplotlib 모듈 호출

X = range(100)
Y = range(100)
Y2 = []
plt.plot(X, Y)
```

```python
import numpy as np # numpy 모듈 호출

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)
plt.show()
```
