```python
import pandas as pd
import numpy as np

raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy', 'Yeonghyun', 'nana', 'gildong'],
            'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze', 'Kim', 'Ann', 'hong' ],
            'age': [42, np.nan, 36, 24, 73, 24, 22, 90],
            'sex': ['m', np.nan, 'f', 'm', 'f', 'm', 'f', 'm'],
            'preTestScore': [4, np.nan, np.nan, 2, 3, 5, 8, 9],
            'postTestScore': [25, np.nan, np.nan, 62, 70, 90, 50, 55]}

df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
df

print(df.fillna(0))



```

```python
#기말고사에 나올 확률 높음.
df["preTestScore"] = df["preTestScore"].fillna(df["preTestScore"].mean())
df
```

```python
group_df = df.groupby('sex')["postTestScore"].transform("mean")
print(group_df)
```

```python
df.loc[2, 'postTestScore'] = group_df[2]
print(df)
```

```python
#실전에서 많이 사용하는 방법
for i in df.index:
    if pd.isnull(df.loc[i, 'postTestScore']):
        df.loc[i, 'postTestScore'] = group_df[i]
print(df)
```

```python
df['postTestScore'] = df['postTestScore'].fillna(df.groupby('sex')['postTestScore'].transform('mean'))
print(df)
```
