```python
# 1부터 100까지 홀수, 짝수 합 구하기
# 방법 1
홀수합 = 0
짝수합 = 0
for i in range(101):
    if i%2 == 0:
        짝수합 += i
    else:
        홀수합 += i
print(f"홀수합: {홀수합}, 짝수합: {짝수합}")
```

```python
홀수합2 = 0
짝수합2 = 0

for i in range(0, 101, 2):
    짝수합2 += i
    홀수합2 += i+1
print(f"홀수합2: {홀수합2}, 짝수합2: {짝수합2}")
```

```python
#등차수열 합을 이용해서 구하기
#1부터 100까지 홀수, 짝수의 합 구하기

홀수합4 = 50*(1+99) // 2
짝수합4 = 50*(2+100) // 2

print(f"홀수합4 : {홀수합4}, 짝수합4: {짝수합4}")
```

```python
#파이썬 스럽게 짜기.
#리스트 함축(List comprehension) = 내포
#리스트 함축 = 리스트를 간결하게 만드는 방법
#무조건 암기!! 중간고사 시험에 낸다!!
#리스트함축 관련해서 시험 문제에 나옴 
#1[표현식 for 항목 in 반복가능객체]
#2[표현식 for 항목 in 반복가능객체 if 조건문]
#조건을 만족하면 값을 추가
#3[표현식 if 조건문 else 표현식 for 항목 in 반복가능객체]
#조건을 만족하면 값, 만족하지 못하면 else에 표현된 값.
#연습, 1부터 100까지 리스트 함축을 이용해서 리스트 만들기

리스트1 = [i for i in range(11)]
리스트2 = [i for i in range(11) if i%2 == 1]
리스트3 = [i if i%2 == 1 else 0 for i in range(11)]
print(리스트1, 리스트2, 리스트3, sep='\n')

홀수리스트1 = [i for i in range(101) if i % 2 != 0]
짝수리스트1 = [i for i in range(101) if i % 2 == 0]
홀수합5 = sum(홀수리스트1)
짝수합5 = sum(짝수리스트1)
```

```python
#중간고사 시험 문제 1부터 100까지의 합을 각각 한줄로 만들기
홀수합6 = sum([i for i in range(101) if i % 2 != 0])
짝수합6 = sum([i for i in range(101) if i % 2 == 0])
print(f"홀수합6: {홀수합6}, 짝수합6: {짝수합6}")
```
