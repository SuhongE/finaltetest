```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#pyplot 기본적인 그래프 그리기 기능 제공하는 클래스! 중요하다 시험 가능(matplotlib의 하위 모듈)
x = np.linspace(0, 10, 10)
print(x)

y1 = 3 * x + 5
y2 = 2 * x**2 + 3 * x + 1



print(y1)


cm_to_inch = 1 / 2.54 # 1cm를 인치로 변환하는 비율
plt.figure(figsize=(10 * cm_to_inch, 6 * cm_to_inch))
#x와y 데이터를 이용해 선 그래프 그리기
#label: 범례에 표시될 이름
plt.plot(x, y1, label='y1 = 3x + 5', color='blue', linewidth=1, linestyle='--',
        marker='s', markersize=5, markerfacecolor='red', markeredgewidth=1, alpha=0.8)
plt.plot(x, y2, label= 'y = 2x^2 + 3x + 1', color='green', linewidth=2, linestyle='-.',marker='s', markersize=5,
         markerfacecolor='yellow', markeredgewidth=1, alpha=0.8)
#범례 표시

#plt.legend() 범례 위치 설정 (upper right, upper left, lower left, lower right, right, left)
plt.legend()
plt.legend(loc='upper right', fontsize=6)

#그래프 제목 설정
#plt.title('Graph of y = 3x + 5')
#제목 폰트 크기 및 굵기 설정
plt.title('Graph of y = 3x + 5', fontsize=10, fontweight='bold')

#x축 라벨 설정, 폰트 크기 및 굵기 설정
plt.xlabel('x', fontsize=14, fontweight='bold')
#y축 라벨 설정, 폰트 크기 및 굵기 설정
plt.ylabel('y', fontsize=10, fontweight='bold')
#y축 라벨 가로 쓰기
plt.ylabel('y축', fontsize=10, rotation=0, labelpad=20)
#한글 폰트 깨짐 방지
plt.rc('font', family='Malgun Gothic')
#격자 표시(True로 설정하면 격자가 보임.)
plt.grid(True)
#격자 스타일 설정(선 스타일, 색상, 두께)
plt.grid()
#그래프 jpg저장하기
plt.savefig('Graph1.jpg', format='jpg', dpi=300)
#그래프 png 저장하기
plt.savefig('graph1.png', format='png', dpi=300)



#그래프 화면에 출력
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

y1 = 3 * x + 5
y2 = 2 * x**2 + 3 * x + 1

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

ax1.plot(x,y1, label='y = 3x + 5', color='blue', linewidth=2)
ax1.set_title('Graph of y = 3x + 5')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)
ax1.legend()

ax2.plot(x, y2, label='y = 2x^2 + 3x + 1', color='green', linewidth=1)
ax2.set_title('Graph of y = 2x^2 + 3x + 1')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)
ax2.legend()

plt.subplots_adjust(
    left=0.1,
    right=0.95,
    top=0.9,
    bottom=0.1,
    wspace=0.2,
    hspace=0.3
)

plt.savefig('graph2.png', format='png', dpi=300)

plt.show()
```

```python
#서브플롯 2행2열 
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0, 10, 100)
y1 = 3 * x + 5
y2 = 2 * x**2 + 3 * x + 1
y3 = np.sin(x)
y4 = np.log(x + 1)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
# 첫 번째 서브플롯
ax1.plot(x, y1, label='y = 3x + 5', color='blue', linewidth=2)
ax1.set_title('Graph of y = 3x + 5')    
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)
ax1.legend()
# 두 번째 서브플롯
ax2.plot(x, y2, label='y = 2x^2 + 3x + 1', color='green', linewidth=2)
ax2.set_title('Graph of y = 2x^2 + 3x + 1')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)
ax2.legend()
# 세 번째 서브플롯
ax3.plot(x, y3, label='y = sin(x)', color='red', linewidth=2)
ax3.set_title('Graph of y = sin(x)')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.grid(True)
ax3.legend()
# 네 번째 서브플롯
ax4.plot(x, y4, label='y = log(x + 1)', color='purple', linewidth=2)
ax4.set_title('Graph of y = log(x + 1)')
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.grid(True)
ax4.legend()

# 그래프 화면에 출력
plt.show()
```

```python
#가로 막대 그래프 그리기 1 bar 2 barh 3 vbar 4 hbar  시험에 나옴!!!

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
# 막대 그래프 그리기
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Graph Example')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.grid(axis='y')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
# 가로 막대 그래프 그리기
plt.figure(figsize=(8, 6))
plt.barh(categories, values, color='salmon')
plt.title('Horizontal Bar Graph Example')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.grid(axis='x')
plt.show()
```

```python
#%% 히스토그램 예시
data = np.random.randn(1000)
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='lightgreen', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
#%% 산점도 예시   x,y를 무작위 난수 생성(random사용) 시험에 나올 수 있음.
x = np.random.rand(100)
y = np.random.rand(100)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='purple', alpha=0.6)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()   
```

```python
#%% 파이 차트 예시
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.axis('equal')  # 원형으로 표시
plt.show()
```
